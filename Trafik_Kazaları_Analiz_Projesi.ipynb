```python
# Gerekli Kütüphane ve Fonksiyonlar
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Trafik akışının değişmesi kazaları nasıl etkiler?

yol_kaza = df["1st_Road_Class"].value_counts().sort_index()

plt.figure(figsize=(8,5))
sns.barplot(x=yol_kaza.index, y=yol_kaza.values)
plt.title("Yol Türüne Göre Kaza Sayısı")
plt.xlabel("1st_Road_Class (1=Motorway, 2=A-road, 3=B-road...)")
plt.ylabel("Kaza Sayısı")
plt.grid(axis='y')
plt.show()
```

```python
trafik_vs_kaza = df.groupby("Year").agg({
    "Number_of_Vehicles": "mean",
    "Accident_Severity": "count"  # her satır bir kaza olduğundan
}).reset_index()

trafik_vs_kaza.columns = ["Yıl", "Ortalama_Araç_Sayısı", "Kaza_Sayısı"]

plt.figure(figsize=(10,6))
plt.plot(trafik_vs_kaza["Yıl"], trafik_vs_kaza["Ortalama_Araç_Sayısı"], label="Ortalama Araç Sayısı", marker='o')
plt.plot(trafik_vs_kaza["Yıl"], trafik_vs_kaza["Kaza_Sayısı"], label="Toplam Kaza Sayısı", marker='s')
plt.xlabel("Yıl")
plt.ylabel("Değer")
plt.title("Yıllara Göre Araç Sayısı ve Kaza Sayısı")
plt.legend()
plt.grid(True)
plt.show()
```

```python
b_road_df = df[df["1st_Road_Class"] == 3]

speed_counts = b_road_df["Speed_limit"].value_counts().sort_index()

plt.figure(figsize=(8,5))
sns.barplot(x=speed_counts.index, y=speed_counts.values, palette="Reds")
plt.title("B-road Kazalarında Hız Limiti Dağılımı")
plt.xlabel("Hız Limiti (mph)")
plt.ylabel("Kaza Sayısı")
plt.grid(axis='y')
plt.show()

rural_urban = b_road_df["Urban_or_Rural_Area"].value_counts()

plt.figure(figsize=(6,4))
sns.barplot(x=rural_urban.index.map({1: "Kentsel", 2: "Kırsal"}), y=rural_urban.values, palette="Greens")
plt.title("B-road Kazalarının Kırsal / Kentsel Dağılımı")
plt.xlabel("Alan")
plt.ylabel("Kaza Sayısı")
plt.show()


```

```python
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M', errors='coerce')

df['Hour'] = df['Time'].dt.hour

# Saatlere göre kaza sayısı
hourly_accidents = df['Hour'].value_counts().sort_index()

plt.figure(figsize=(12,5))
hourly_accidents.plot(kind='bar')
plt.title("Saatlere Göre Kaza Sayısı")
plt.xlabel("Saat")
plt.ylabel("Kaza Sayısı")
plt.show()
```

```python
#Kaza oranlarını ne artırır?

df.groupby('Weather_Conditions')['Accident_Index'].count().sort_values(ascending=False)
df.groupby('Road_Type')['Accident_Index'].count()
df.groupby('Light_Conditions')['Accident_Index'].count()
```

```python
weather_counts = df.groupby('Weather_Conditions')['Accident_Index'].count().sort_values(ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(y=weather_counts.index, x=weather_counts.values, palette="Blues_r")
plt.title("Hava Koşullarına Göre Kaza Sayısı")
plt.xlabel("Kaza Sayısı")
plt.ylabel("Hava Durumu")
plt.tight_layout()
plt.show()
```

```python
sns.countplot(data=df, x="Light_Conditions", hue="Accident_Severity")
plt.xticks(rotation=45)
plt.title("Işık Koşullarına Göre Kaza Şiddeti Dağılımı")
plt.show()
```

```python
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
df["Year"] = df["Date"].dt.year

df["Year"].value_counts().sort_index().plot(kind="bar", figsize=(12,6), title="Yıllara Göre Kaza Sayısı")
plt.xlabel("Yıl")
plt.ylabel("Kaza Sayısı")
plt.show()
```

```python
#Zaman içinde kaza oranlarını tahmin edebilir miyiz?

df4=pd.read_csv('ukTrafficAADF.csv')

df_4=df4.drop(['AADFYear', 'CP','Region', 'LocalAuthority', 'Road', 'RoadCategory', 'Easting',
       'Northing', 'StartJunction', 'EndJunction', 'LinkLength_km',
       'LinkLength_miles', 'PedalCycles', 'Motorcycles', 'CarsTaxis',
       'BusesCoaches', 'LightGoodsVehicles', 'V2AxleRigidHGV',
       'V3AxleRigidHGV', 'V4or5AxleRigidHGV', 'V3or4AxleArticHGV',
       'V5AxleArticHGV', 'V6orMoreAxleArticHGV', 'AllHGVs', 'AllMotorVehicles',
       'Lat', 'Lon'],axis=1)

df_4.isna().sum()
df_4.dropna(inplace=True)

estimation_array=df_4["Estimation_method"].unique()
estimation=df_4["Estimation_method"].value_counts().values
plt.pie(estimation,labels=estimation_array,radius=2,startangle=25,textprops={"size":"medium"},autopct='%1.1f%%',explode=(0.02,0.02))
plt.title("Estimation method of predicted accidents")
plt.show()


estimation_method_array=df_4["Estimation_method_detailed"].unique()
estimation_method=df_4["Estimation_method_detailed"].value_counts().values
plt.pie(estimation_method,labels=estimation_method_array,radius=2,startangle=25,textprops={"size":"medium"},autopct='%1.1f%%',explode=(0.02,0.02,0.02,0.02,0.02))
plt.title("The way predictions are made")
plt.show()
```

```python
#Kırsal ve kentsel alanlar nasıl farklılaştı?

df['Date'] = pd.to_datetime(df['Date'], errors='coerce')


df = df[df['Date'].notna()]

df['Year'] = df['Date'].dt.year

urban_counts = df[df['Urban_or_Rural_Area'] == 1].groupby('Year').size()
rural_counts = df[df['Urban_or_Rural_Area'] == 2].groupby('Year').size()

plt.plot(urban_counts.index, urban_counts.values, label='Kentsel')
plt.plot(rural_counts.index, rural_counts.values, label='Kırsal')
plt.title('Yıllara Göre Kentsel ve Kırsal Alanlarda Kaza Sayısı')
plt.xlabel('Yıl')
plt.ylabel('Kaza Sayısı')
plt.legend()
plt.show()
```
